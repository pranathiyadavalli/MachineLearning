```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('/content/drive/MyDrive/Mall_Customers.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'])
plt.show()
```

```python
X = data.iloc[:,[3,4]].values
```

```python
hier_model = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
```

```python
y_hier_model = hier_model.fit_predict(X)
```

```python
y_hier_model
```

```python

plt.figure(figsize=(12,7))
plt.scatter(X[y_hier_model == 0, 0], X[y_hier_model == 0, 1], s = 100, c = 'red', label = 'Careful')
plt.scatter(X[y_hier_model == 1, 0], X[y_hier_model == 1, 1], s = 100, c = 'blue', label = 'Standard')
plt.scatter(X[y_hier_model == 2, 0], X[y_hier_model == 2, 1], s = 100, c = 'green', label = 'Target group')
plt.scatter(X[y_hier_model == 3, 0], X[y_hier_model == 3, 1], s = 100, c = 'orange', label = 'Careless')
plt.scatter(X[y_hier_model == 4, 0], X[y_hier_model == 4, 1], s = 100, c = 'magenta', label = 'Sensible')
plt.title('Clustering of customers',fontsize=20)
plt.xlabel('Annual Income (k$)',fontsize=16)
plt.ylabel('Spending Score (1-100)',fontsize=16)
plt.legend(fontsize=16)
plt.grid(True)
plt.axhspan(ymin=60,ymax=100,xmin=0.4,xmax=0.96,alpha=0.3,color='yellow')
plt.show()
```

```python

import scipy.cluster.hierarchy as sch

plt.figure(figsize=(15,6))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
#plt.grid(True)

dendrogram = sch.dendrogram(sch.linkage(X,method='ward'), truncate_mode='lastp', p=10)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.hlines(y=190,xmin=0,xmax=2000,lw=3,linestyles='--')
plt.text(x=900,y=220,s='Horizontal line crossing 5 vertical lines',fontsize=20)
#plt.grid(True)
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.show()
```
